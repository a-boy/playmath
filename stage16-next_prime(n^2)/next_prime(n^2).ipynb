# next_prime(n^2)

改进Oppermann猜想：next_prime(n^2) - n^2 <= 1+euler_phi(n)



NextPrime-Square-Inequality;for n>0, **`next_prime(n^2) - n^2 <= 1+euler_phi(n)`**

奥珀曼猜想 Oppermann's conjecture



```python
def next_prime_delta(x):
    return next_prime(x)-x
def previous_prime_delta(x):
    return previous_prime(x)-x
```
