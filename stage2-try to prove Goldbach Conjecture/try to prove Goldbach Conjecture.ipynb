**Improved Goldbach's Conjecture with prime_pi bounded**:
For any integer n>2, there exists a integer j, 0<=j<=prime_pi(n)+1 making both n-j and n+j be prime. Here prime_pi is the prime count function.

**define**: golddelta(n) is the least non-negative integer makes both n-golddelta(n) and n+golddelta(n) be prime

the below code verifies 0<=golddelta(n)<=prime_pi(n)+1

```python
def golddelta(n):
    j=0
    while not(is_prime(n-j) and is_prime(n+j) ):
        j+=1
    return j
li=[(n,golddelta(n),prime_pi(n),floor(2*sqrt(n))) for n in range(3,1000)]
print(li)
```
