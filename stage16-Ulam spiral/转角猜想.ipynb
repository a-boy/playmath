质数沉默的表示，在下一个转角再见！

```python
from sympy import isprime 
```

```python
for n in range(1,100):
    tri=n*n-n+1,n*n+1,n*n+n+1
    
    if not isprime(tri[0]) and not isprime(tri[1]) and not isprime(tri[2]):
        print(n,tri)
```

```python
li=[n for n in range(2**100,2**100+2000) if isprime(n*n+1) ]
print(li)
m=li[0]
gap=[]
for n in li[1:]:
    gap.append(n-m)
    m=n
print(gap)
    
```

```python
for n in range(1,10**3):
    
    t=n*n-n-1,n*n-n+1, n*n-2, n*n+1,  n*n+n-1,n*n+n+1, (n+1)**2-2,(n+1)**2+1
    print(n,any(map(isprime,t)))
```
