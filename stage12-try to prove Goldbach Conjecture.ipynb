**Improved Goldbach's Conjecture**  : For any positive integer `n>1`, there exists `g, 0=<g<=prime_pi(n)+1`, make that both `n-g` and `n+g` are primes.
`2*n=(n-g)+(n+g)`  



```python
def golds(n):
    return [g for g in range(0,prime_pi(n)+2) if (is_prime(n-g) and is_prime(n+g))]

def gold(n):
    for g in range(0,prime_pi(n)+2):
        if (is_prime(n-g) and is_prime(n+g)):
            return g

golds_list=[golds(n) for n in range(2,256)]
print(golds_list)
```

```python
# check Improved Goldbach's Conjecture
n=2
while gold(n)>=0:
    n+=1
```
